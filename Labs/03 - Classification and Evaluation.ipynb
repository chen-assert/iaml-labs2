### ========== Question 2.9 ==========
Now we want to explore the effect of the `k` parameter. To do this, train the classifier multiple times, each time setting the KNN option to a different value. ### ========== Question 2.11 ==========
Select best value for `k` from Questions 2.9 and 2.10 and plot the normalised confusion matrix on the test set (you may use the provided function). Then plot the confusion matrix for a 5-nearest neighbour classifier. What do you observe?

```python
def plot_confusion_matrix(cm, classes=None, title='Confusion matrix'):
    """Plots a confusion matrix."""
    if classes is not None:
        sns.heatmap(cm, xticklabels=classes, yticklabels=classes, vmin=0., vmax=1., annot=True)
    else:
        sns.heatmap(cm, vmin=0., vmax=1.)
    plt.title(title)
    plt.ylabel('True label')
```

```python
# Your Code goes here:
neigh = KNeighborsClassifier(n_neighbors=1000)
neigh.fit(X_train, y_train) 
cm = confusion_matrix(y_test, neigh.predict(X_test))
cm_norm = cm/cm.sum(axis=1)[:, np.newaxis]
plot_confusion_matrix(cm_norm,classes=neigh.classes_)
```

```python
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(X_train, y_train) 
cm = confusion_matrix(y_test, neigh.predict(X_test))
cm_norm = cm/cm.sum(axis=1)[:, np.newaxis]
plot_confusion_matrix(cm_norm,classes=neigh.classes_)
```
